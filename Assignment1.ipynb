```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
numerical_columns = [
    "Volume",
    "Avg Price",
    "Total Sales Value",
    "Discount Rate (%)",
    "Discount Amount",
    "Net Sales Value",
]
```

```python
df_standardized = df.copy()
df_standardized[numerical_columns] = sc.fit_transform(df[numerical_columns])
```

```python
print("Standardized Data:")
display(df_standardized.head())
```

```python
categorical_columns = ["SKU", "City", "Brand", "Model"]
df_encoded = pd.get_dummies(df_standardized, columns=categorical_columns, drop_first=True)
```

```python
print("One-Hot Encoded Data:")
display(df_encoded.head())
```

```python
df_encoded.to_csv("transformed_dataset.csv", index=False)
```
